```python
# Question 1 )  What is a Convolutional Neural Network (CNN), and why is it used for image processing?

# A CNN is a type of deep learning model specialized for grid-like data, such as images. It automatically and adaptively learns spatial hierarchies of features through convolutional layers, making it ideal for image recognition, classification, and detection tasks.




# Question 2 )  What are the key components of a CNN architecture?

# Input Layer – Image data (height × width × channels)

# Convolutional Layers – Extract features

# Activation Functions (ReLU) – Introduce non-linearity

# Pooling Layers – Reduce dimensionality

# Fully Connected Layers – Perform classification

# Output Layer – Produces final prediction (e.g., class probabilities)




# Question 3)  What is the role of the convolutional layer in CNNs?

# It applies a set of filters (kernels) that slide over the input image to extract features like edges, corners, textures, etc., producing feature maps.



#Question 4️)  What is a filter (kernel) in CNNs?

# A small matrix (e.g., 3×3 or 5×5) of trainable weights that performs element-wise multiplication and summation with the input region, helping the network learn patterns.




# Question 5️)  What is pooling in CNNs, and why is it important?

# Pooling reduces the spatial dimensions of feature maps (height and width), retaining the most important information. It:

# Reduces computational cost

# Prevents overfitting

# Introduces spatial invariance




# Question 6️)  What are the common types of pooling used in CNNs?

# Max Pooling – Selects the maximum value from a region

# Average Pooling – Computes the average value from a region

# Global Average Pooling – Reduces the entire feature map to a single value per channel




# Question 7️) How does the backpropagation algorithm work in CNNs?


# Backpropagation in CNNs computes the gradient of the loss function w.r.t. each weight using the chain rule, then updates the weights using gradient descent. It handles convolutional and fully connected layers differently due to weight sharing in convolutions.



# Question 8)  What is the role of activation functions in CNNs?

# They introduce non-linearity so that the model can learn complex functions. Common activation functions:

# ReLU – f(x) = max(0, x)

# Sigmoid – Good for binary output

# Softmax – For multi-class classification (in the output layer)



# Question 9️)  What is the concept of receptive fields in CNNs?
# The receptive field is the region in the input image that a particular neuron in a feature map "sees." As you move deeper in the network, receptive fields grow larger, capturing more global context.



# Question 10)  Explain the concept of tensor space in CNNs?

# A tensor is a multi-dimensional array (generalized matrix). In CNNs, data flows through as:

# 3D tensor: (Height × Width × Channels) for images

# 4D tensor: (Batch × Height × Width × Channels) during training

# This structure helps process and compute over batches efficiently.



# Question 11)  What is LeNet-5, and how does it contribute to the development of CNNs?

# Developed by Yann LeCun in the 1990s for handwritten digit recognition (MNIST).

# Introduced convolution + pooling + fully connected layers.

# One of the earliest CNNs, forming the foundation for modern architectures.



# Question 12)  What is AlexNet, and why was it a breakthrough in deep learning?

# Won ImageNet 2012 challenge with huge margin.

# Used ReLU, dropout, data augmentation, and GPU training.

# Deep network (8 layers) — brought CNNs to mainstream in computer vision.



#  Question13)  What is VGGNet, and how does it differ from AlexNet?


# Uses only 3×3 convolution filters and very deep architecture (16 or 19 layers).

# Simpler, more uniform design compared to AlexNet.

# Better performance with deeper yet cleaner structure.




# Question 14)   What is GoogLeNet, and what is its main innovation?

# Introduced Inception Modules — multiple filter sizes in one layer.

# Reduced parameters using 1×1 convolutions (bottleneck layers).

# Won ImageNet 2014. Very efficient and deep.




# Question 15)  What is ResNet, and what problem does it solve?


# Introduced residual connections (skip connections).

# Solves vanishing gradient problem in very deep networks.

# Enabled training of ultra-deep networks (e.g., 152 layers).




#  Question 16)  What is DenseNet, and how does it differ from ResNet?

# In DenseNet, each layer is connected to all subsequent layers.

# Promotes feature reuse, reduces parameters.

# Unlike ResNet, which adds output of layers, DenseNet concatenates them.




# Questioin 17)   What are the main steps involved in training a CNN from scratch?


# Prepare dataset – Images + labels, resize/normalize, split into train/val/test

# Define model architecture – Layers, activations, optimizer

# Forward pass – Compute predictions

# Loss computation – Measure error using a loss function

# Backpropagation – Calculate gradients

# Update weights – Using an optimizer like SGD or Adam

# Repeat (epochs) – Until model converges

# Evaluate – Accuracy, loss, confusion matrix

# Deploy – Save and use the trained model

# If you want this in a PDF or formatted notes file, let me kn
```

```python
# Question 1)  Implement a basic convolution operation using a filter and a 5x5 image (matrix) 
import numpy as np
from scipy.signal import convolve2d

image = np.array([
    [1, 2, 3, 0, 1],
    [4, 5, 6, 1, 2],
    [7, 8, 9, 0, 1],
    [1, 3, 2, 2, 3],
    [0, 1, 2, 3, 4]
])

kernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])

convolved = convolve2d(image, kernel, mode='valid')
print("Convolved Output:\n", convolved)
```

```python
# Question 2)  Implement max pooling on a 4x4 feature map with a 2x2 window

import numpy as np
from skimage.measure import block_reduce

feature_map = np.array([
    [1, 3, 2, 1],
    [4, 6, 5, 2],
    [7, 2, 8, 3],
    [1, 0, 6, 4]
])

pooled = block_reduce(feature_map, block_size=(2, 2), func=np.max)
print("Max Pooled Output:\n", pooled)
```

```python
# Question 3) Implement the ReLU activation function on a feature map

feature_map = np.array([
    [-1, 2],
    [3, -4]
])

relu_activated = np.maximum(0, feature_map)
print("ReLU Output:\n", relu_activated)
```

```python
# Question 4 ) Create a simple CNN model with one convolutional layer and a fully connected layer, using random data 

import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(8, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.summary()
```

```python
# Question 5)  Generate a synthetic dataset using random noise and train a simple CNN model on it


import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense

X = np.random.rand(100, 28, 28, 1)
y = tf.keras.utils.to_categorical(np.random.randint(0, 10, 100), 10)

model = Sequential([
    Conv2D(16, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    Flatten(),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=5, batch_size=16)

```
